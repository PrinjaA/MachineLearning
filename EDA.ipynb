```python
df = df[['coaster_name',
         'Location',
         'Status',
         'Manufacturer',	
         'year_introduced',
         'latitude',
         'longitude',
         'Type_Main',
         'opening_date_clean',
         'speed_mph',
         'height_ft',
         'Inversions_clean',
         'Gforce_clean']].copy()
```

```python
df
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df = df.loc[~df.duplicated(subset=['coaster_name','Location','opening_date_clean'])].reset_index(drop=True)
```

```python
df['year_introduced'].value_counts()
```

```python
ax = df['year_introduced'].value_counts().head(10) \
     .plot(kind='bar', title='Top 10 coasters Introduced')
ax.set_xlabel('year introduced')
ax.set_ylabel('count')
```

```python
ax = df['speed_mph'].plot(kind='hist',bins=20, title='coaster speed(mph)')
ax.set_xlabel('speed mph')
```

```python
ax = df['speed_mph'].plot(kind='kde', title='coaster speed(mph)')
ax.set_xlabel('speed mph')
```

```python
df.plot(kind='scatter', x='speed_mph', y='height_ft', title='Coaster Speed VS Height')
plt.show()
```

```python
ax = sns.scatterplot(x='speed_mph', y='height_ft', hue='year_introduced', data=df)
ax.set_title('Coaster Speed VS Height')
plt.show()
```

```python
sns.pairplot(df,vars=['year_introduced','speed_mph',
       'height_ft', 'Inversions_clean', 'Gforce_clean'], hue='Type_Main')
plt.show()
```

```python
df_corr = df[['year_introduced','speed_mph',
       'height_ft', 'Inversions_clean', 'Gforce_clean']].dropna().corr()
df_corr
```

```python
sns.heatmap(df_corr,annot=True)
```
